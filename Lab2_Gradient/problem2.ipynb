### Oracles

Numerical optimization methods need callable *oracles* for properties of $f$, that is a function that, given a point $x$ in the domain of $f$, returns $f$ and/or gradient, Hessian of $f$ at point $x$. We talk about the *order* of an oracle as the number of differentiations given (0th order for just $f$, 1st order for the gradient, 2nd for gradient + Hessian).

> Observe the first order oracle `f_grad`.


```python
import numpy as np

##### Gradient oracle
def f_grad(x):
    x1 = x[0]
    x2 = x[1]
    gx = 8*(x1-3)*np.exp(4*(x1-3)**2)/( 1 + np.exp(4*(x1-3)**2) + np.exp(2*(x2-1)**2) )
    gy = 4*(x2-1)*np.exp(2*(x2-1)**2)/( 1 + np.exp(4*(x1-3)**2) + np.exp(2*(x2-1)**2) )
    return np.array( [ gx  ,  gy  ] )
####
```

> Fill the following second order oracle `f_grad_hessian`.

```python
import numpy as np

##### Hessian scaled Gradient computation
def f_grad_hessian(x):
    x1 = x[0]
    x2 = x[1]
    gx = 8*(x1-3)*np.exp(4*(x1-3)**2)/( 1 + np.exp(4*(x1-3)**2) + np.exp(2*(x2-1)**2) )
    gy = 4*(x2-1)*np.exp(2*(x2-1)**2)/( 1 + np.exp(4*(x1-3)**2) + np.exp(2*(x2-1)**2) )
    g = np.array( [ gx  ,  gy  ] )
    dxx = (8*np.exp(4*(x1-3)**2) + (8*(x1-3))**2*np.exp(4*(x1-3)**2) )/( 1 + np.exp(4*(x1-3)**2) + np.exp(2*(x2-1)**2) ) - gx**2
    dxy = -( 8*(x1-3)*np.exp(4*(x1-3)**2)*4*(x2-1)*np.exp(2*(x2-1)**2) )/( 1 + np.exp(4*(x1-3)**2) + np.exp(2*(x2-1)**2) )**2
    dyx = dxy
    dyy = (4*np.exp(2*(x2-1)**2) +  (4*(x2-1))**2*np.exp(2*(x2-1)**2) )/( 1 + np.exp(4*(x1-3)**2) + np.exp(2*(x2-1)**2) ) - gy**2
    H = np.array(  [ (  dxx   , dxy )  ,  ( dyx , dyy )  ]  )
    return g,H
####
```
